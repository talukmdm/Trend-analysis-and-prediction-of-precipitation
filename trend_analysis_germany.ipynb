```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np
all_files=glob.glob('*.txt')
li=[]
for i in all_files:
    df = pd.read_csv(i,sep=";",skiprows=1)
    j=df.drop(["Deutschland","Unnamed: 19"],axis=1)
    column=list(j.columns[2:])
    f=pd.melt(df,id_vars=["Monat","Jahr"],value_vars=column,var_name="States",value_name=str(j.loc[1][1]))
    li.append(f)

    
from functools import reduce
#df_merged = reduce(lambda  left,right: pd.merge(left,right,how='outer'),li)
data=pd.DataFrame({"YEAR":li[0]["Jahr"],"State":li[1]["States"],"JAN":li[0]["1.0"],
                    "FEB":li[1]["2.0"],"MAR":li[2]["3.0"],"APR":li[3]["4.0"],"MAY":li[4]["5.0"],
                   "JUN":li[5]["6.0"],"JUL":li[6]["7.0"],"AUG":li[7]["8.0"],"SEP":li[8]["9.0"],
                    "OCT":li[9]["10.0"],"NOV":li[10]["11.0"],"DEC":li[11]["12.0"]})
```

# adding seasonal and annual columns


```python
data["DEC-FEB"]=data.iloc[:,2:4].sum(axis=1)
data["MAR-MAY"]=data.iloc[:,4:7].sum(axis=1)
data["JUN-AUG"]=data.iloc[:,7:11].sum(axis=1)
data["SEP-NOV"]=data.iloc[:,11:14].sum(axis=1)
data["ANNUAL"]=data.iloc[:,2:14].sum(axis=1)
data
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pymannkendall as mk
def STATES(column,division):
        Year = []
        annual = []
        for i in range(len(column)):
            if column[i] == division:
                if 'nan'!= str(data["ANNUAL"][i]) and 'NAN' != str(data["ANNUAL"][i]):
                    Year.append(data["YEAR"][i])
                    annual.append((data["ANNUAL"][i]))
        return Year,annual
colors = ["#808000","#FF0000","#0000FF","#808000","#800080","#008000","#800000","#A52A2A","#FFA500","#000000","#151B54","#FBB917","#806517","#C11B17",
          "#810541","#F6358A","#808000","#FF0000","#0000FF","#808000","#800080","#008000","#800000","#A52A2A","#FFA500","#000000","#151B54","#FBB917",
          "#806517","#C11B17","#810541","#F6358A","#808000","#FF0000","#0000FF","#808000"]
country = data["State"].unique()
name=list(data["State"].unique())
for j in range(len(country)):
    Year, annual = STATES(data["State"],str(country[j]))
    res = mk.original_test(annual,alpha=.05)
    print(res)
    trend_line = np.arange(len(annual)) * res.slope + res.intercept
    plt.style.use('fivethirtyeight')
    plt.rcParams['figure.figsize']=(25,5)
    fix,ax=plt.subplots(1,1)
    ax.plot(Year, annual, color=colors[j] ,alpha = 0.7, marker='o', linestyle='dashed',linewidth=2, markersize=12)
    ax.plot(Year, trend_line)
    plt.title(str(country[j])+" Annual Rainfall\n", size= 25)
    plt.xlabel("Year", size= 25)
    plt.ylabel("Rainfall in\nmm", size= 30)
    plt.tight_layout()
    x=country[j].replace('/', '_')
   # plt.savefig("plots/"+x+".jpg",dpi=150)
    plt.show()
```

# Change point anallysis by Pettitt test

```python
import pyhomogeneity as hg
cp_data = data.groupby(["State"])["ANNUAL","YEAR"]
cp_df=pd.DataFrame(columns=["states","Change Point","p","h"],index=(range(16)))
states=data["State"].unique()
for q in range(len(country)):
    state=country[q]
    grp=cp_data.get_group(state)
    grp=grp.set_index("YEAR")
    grp.index=pd.to_datetime(grp.index,format="%Y")
    result = hg.pettitt_test(grp,alpha=0.05)
    cp_df["states"][q]=state
    cp_df["Change Point"][q]=result.cp[0:4]
    cp_df["p"][q]=result.p
    cp_df["h"][q]=result.h
cp_df.to_csv("Pettit test annual.csv")
```

# Train test splitiing

```python
#tarin tes split for the year 2010
from sklearn.model_selection import train_test_split

data_2010 = np.asarray(data[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].loc[(data['YEAR'] >= 2010) & (data["YEAR"] <=2015)])

#data_2010 = np.asarray(data[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       #'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].loc[(data['YEAR'] == 2010)])

#data_2010_brandenburg = np.asarray(data_2010[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       #'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].loc[data_2010['State'] == 'Brandenburg/Berlin'])

X_year_2010 = None; y_year_2010 = None
for i in range(data_2010.shape[1]-3):
    if X_year_2010 is None:
        X_year_2010 = data_2010[:, i:i+3]
        y_year_2010 = data_2010[:, i+3]
    else:
        X_year_2010 = np.concatenate((X_year_2010, data_2010[:, i:i+3]), axis=0)
        y_year_2010 = np.concatenate((y_year_2010, data_2010[:, i+3]), axis=0)
        
X_train, X_test, y_train, y_test = train_test_split(X_year_2010,y_year_2010,test_size = 0.20,random_state = 5)
```

```python
###### knn model for 2010 data
%matplotlib
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import accuracy_score
neigh = KNeighborsRegressor(n_neighbors=40)
neigh.fit(X_train,y_train)

predicted_knn = neigh.predict(X_test)
#print("Prediction Result: ",predicted1)
mean_abs_per = mean_absolute_percentage_error(predicted_knn,y_test)
#print('\nR-squared test score: {:.3f}'.format(neigh.score(X_test_i,y_test_i))) # R-Squared test score
plt.rcParams['figure.figsize']=(16,4)
plt.plot(y_test,label='Actual')
plt.plot(predicted_knn,label='Predicted by k-NN model')
plt.xlabel('Test Size',alpha=0.8,fontsize=20)
plt.ylabel('Precipitation ( in mm)',alpha=0.8,fontsize=20)
plt.xticks(alpha=0.6,fontsize=20)
plt.yticks(alpha=0.6,fontsize=20)
plt.legend(fontsize=20,loc=1)
plt.text(100,175,"MAPE: "+str(mean_abs_per.round(3)),fontsize=20,horizontalalignment='center')
#plt.text(12,125,"MAPE: "+str(mean_abs_per.round(3)),fontsize=20,horizontalalignment='center')
plt.show()
plt.tight_layout()
#plt.savefig("knn_k20.jpg",dpi=150)

#from sklearn.utils import check_arrays
#mean_abs = mean_absolute_error(predicted_knn,y_test)

print(mean_abs_per)

#plt.savefig("plots/knn_40.jpg",dpi=150)
```
